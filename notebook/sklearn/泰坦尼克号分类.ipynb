```python
path = "http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic.txt"
df = pd.read_csv(path)
```

```python
y = df["survived"]
x = df [["pclass","age","sex"]]
```

```python
x["age"].fillna(x["age"].mean(),inplace=True)
```

```python
#转成字典
data = x.to_dict(orient="records")
data
```

```python
#拆分

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(data,y,random_state=42)
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
```

```python
transfer = DictVectorizer()
transfer2 = KNeighborsClassifier()
mypipeline = Pipeline([('t1',transfer),('t2',transfer2)])
```

```python
mypipeline.get_params().keys()
```

```python
mypipeline.set_params(t1__sparse=False,t2__n_neighbors=3)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param = {"t2__n_neighbors":[4,5,6,7,9]}
grid_search = GridSearchCV(mypipeline,param_grid=param,cv=10)
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
y_predict = grid_search.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
myresult = classification_report(y_predict,y_test)
```
