```python
import numpy as np
import pandas as pd
```

```python
path = '../../../MyDataSets/分类数据集/动漫分类/动漫.csv'
df = pd.read_csv(path,names=["x1","x2","x3","x4","x5"])
```

```python
df.head()
```

```python
df = df.dropna()
df = df[:100]
x_data = df.x5
y_data = df.x2
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data,y_data,random_state=22)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import jieba
#中文需要自己分词，然后再进行文本提取
def cuttext(text):
    return " ".join(list(jieba.cut(text)))
```

```python
#先分词
def dealText(data):
    temp = []
    for s in data:
        temp.append(cuttext(str(s)))
    return temp
```

```python
x_train_deal = dealText(x_train)
```

```python
x_train
```

```python
#再文本提取
tranfer = TfidfVectorizer()

x_train_new = tranfer.fit_transform(x_train_deal)
x_test_new = tranfer.transform(x_test)
```
