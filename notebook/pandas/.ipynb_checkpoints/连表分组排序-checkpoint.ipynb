```python
import numpy as np
import pandas as pd
from pandas import read_csv
```

```python
path = "D:\my\datasets\\"
job_path = path + "JOB.csv"
jobad_path = path + "JOBAD.csv"
channel_path = path + "CHANNEL.csv"

jobtable = pd.read_csv(job_path)
jobadtable = pd.read_csv(jobad_path)
channeltable = pd.read_csv(channel_path)
```

 ## 一、表中单列操作

```python
jobtable.columns
```

```python

newcolums_job = [ 'TenantId', 'JobId','JobTitle']
new_jobtable = jobtable[newcolums_job]

newcolums_jobad = [ 'TenantId', 'UserId', 'JobId','JobAdId','JobAdName','AdChannelType','CreateDate']
new_jobadtable = jobadtable[newcolums_jobad]

newcolums_channel = [ 'TenantId', 'ChannelId','ChannelName']
new_channeltable = channeltable[newcolums_channel]
```

```python
df1 = pd.merge(new_jobtable,new_jobadtable,how="left",left_on=['TenantId', 'JobId'],right_on=['TenantId', 'JobId'])
```

```python
df2 = pd.merge(df1,new_channeltable,how='left',left_on=['TenantId', 'AdChannelType'],right_on=['TenantId', 'ChannelId'])
```

```python
df2.dropna(inplace=True)
```

```python
grouped = df2.groupby("ChannelId")["JobAdId"]
```

```python
grouped.count().sort_values(ascending=False)
```

### 思考题1：筛选出发布超过6条广告的所有职位？

```python
df = pd.merge(new_jobtable,new_jobadtable,how="left",left_on=['TenantId', 'JobId'],right_on=['TenantId', 'JobId'])
```

```python
df.dropna(inplace=True)
```

```python
#1.分组
temp = df.groupby(["JobId"]).count()

#2.找到符合条件的职位ID
temp = temp[temp["JobAdId"]>6]

#3.在列表中展示出这些职位：isin
df = df[df.JobId.isin(temp.index)]
```

### 思考题2：筛选出广告发布超过3个渠道的所有职位？

```python
df1 = pd.merge(new_jobtable,new_jobadtable,how="left",left_on=['TenantId', 'JobId'],right_on=['TenantId', 'JobId'])

df2 = pd.merge(df1,new_channeltable,how="left",left_on=['TenantId', 'AdChannelType'],right_on=['TenantId', 'ChannelId'])

df = df2.dropna()
```

```python
#l = list(temp.index)
a1 = temp.index.get_level_values(0)
a2 = temp.index.get_level_values(1)
#list(a)


#temp.index.get_level_values(1)
s1 = pd.Series(a1)
s2 = pd.Series(a2)
s3 = pd.DataFrame([s1,s2])
```
